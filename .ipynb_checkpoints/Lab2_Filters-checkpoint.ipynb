## Exercice 1 : Quelques manipulations simples avec des histogrammes

L'objectif de cet exercice est de refaire quelques manipulations simples d'images au travers de la notion d'histogramme. Vous travaillerez sur une image de votre choix.

+ Ouvrir l'image et la visualiser. On travaillera dans un premier temps sur l'image en niveau de gris.


```python
%matplotlib inline

import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

def loadImage(path):
    if not os.path.exists(path):
        raise FileNotFoundError("img %s does not exist".format(path))
    img=cv2.imread(path,-1)
    # if img.shape[2] == 3:
        # img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img, img.shape

img, shape = loadImage('./Images/oscar.jpg')
plt.imshow(img, cmap="Greys")
plt.show()
```

+ Calculer et visualiser l'histogramme de cette image. 


```python
from cv2 import calcHist

def calculhisto(I):
    if len(I.shape) > 2:
        r = calcHist([I], [0], None, [256], [0,256])
        g = calcHist([I], [1], None, [256], [0,256])
        b = calcHist([I], [2], None, [256], [0,256])
        return (r,g,b)
    else:
        h = calcHist([I], [0], None, [256], [0, 256])
        return h

x = np.linspace(0,255, 256)
rgbHist = calculhisto(img)
plt.figure(figsize=(15,6))
plt.plot(x, rgbHist)
plt.show()
```

+ Ecrire une fonction qui permet de d√©terminer le niveau de gris minimum et le niveau de gris maximum d'une image et l'appliquer √† votre image.


```python
def MinMaxGray(I):
    max = 0
    min = 256
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            if I[i,j] < min:
                min = I[i,j]
            if I[i,j] > max:
                max = I[i,j]
    return min, max

MinMaxGray(img)
    
    
```

 + Ecrire une fonction **egalisationHisto (I,h)** qui, √† partir d'une image $I$ et de son histogramme $h$ renvoie l'image apr√®s √©galisation. Vous pourrez aussi afficher l'histogramme de l'image apr√®s √©galisation.
 
Dans une premi√®re version, vous n'utiliserez pas les m√©thodes propos√©es par la biblioth√®que OpenCV autre que celles permettant de construire un histogramme **calcHist**. Il faudra donc calculer l'histogramme cumul√© et appliquer la transformation vue en cours.


```python
def cumulHisto(h, I):
    cumulHisto = []
    numPixels = I.shape[0]*I.shape[1]
    for i in range(len(h)):
        if i == 0:
            cumulHisto.append(h[i]/numPixels)
        else:
            cumulHisto.append(cumulHisto[i-1] + h[i]/numPixels)
    return cumulHisto

cHisto = cumulHisto(rgbHist, img)
plt.figure(figsize=(15,6))
plt.plot(x, cHisto)
plt.show()
```

```python
def egalisationHisto(h,I):
    cHist = cumulHisto(h, I)
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            I[i,j] = (255*cHist[I[i,j]]).round()
    return I

eqImg = egalisationHisto(rgbHist, img)
plt.imshow(eqImg, cmap="Greys")
plt.show()
```

```python
hist = calculhisto(eqImg)
plt.figure(figsize=(15,6))
plt.plot(x, hist)
plt.show()
```

```python
cumulHist = cumulHisto(hist, eqImg)
plt.figure(figsize=(15,6))
plt.plot(x, cumulHist)
plt.show()
```

```python
import numpy as np

img, shape = loadImage('./Images/oscar.jpg')

def cumulHistoBis(I):
    hist,bins = np.histogram(I.flatten(),256,[0,256])
    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max()/ cdf.max()
    return cdf_normalized

def egalisationHistobis(h,I):
    equ = cv2.equalizeHist(I)
    return equ

eqImg = egalisationHistobis("", img)
plt.imshow(eqImg, cmap="Greys")
plt.show()

cHisto = cumulHistoBis(eqImg)
plt.plot(x, cHisto)
plt.show()
```
